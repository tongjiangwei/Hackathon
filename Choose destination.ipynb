```python
import pandas as pd
df = pd.read_csv('..\\Desktop\\urban data\\Urban Data Hackathon\\UTSEUS-MOBIKE-shanghai_full.csv')
```

```python
df.columns
```

```python
start_location_x = input("start_location_x: ")
start_location_y = input("start_location_y: ")
```

```python

def best_location(start_location_x, start_location_y):
    return df[(df['start_location_x'] == float(start_location_x) ) & (df['start_location_y']  == float(start_location_y))  ].groupby(['end_location_x', 'end_location_y'])['userid'].count().sort_values(ascending=False).reset_index().ix[0]

best_location(start_location_x, start_location_y)
```

```python
print(df.shape)
A = df[['start_location_x', 'start_location_y']].join(df[:500].dropna().apply(lambda r: best_location(r['start_location_x'], r['start_location_y']),1).rename(columns=dict(userid='recommendations'))).sort_values(by='recommendations', ascending=False)
```

```python
A
```

```python
from geojson import LineString, Feature, FeatureCollection
import geojson

A_cut = A[:300]
featureCollection = []
print(dir(featureCollection))
for row in A_cut.values:
    line = LineString([(row[0], row[1]), (row[2], row[3])])
    feature = Feature(geometry = line, properties={"recommendations": row[4]})
    featureCollection.append(feature)

featureCollection = FeatureCollection(featureCollection)
```

```python
geojson.dump(featureCollection, open('geo4.json', "w"))
```
